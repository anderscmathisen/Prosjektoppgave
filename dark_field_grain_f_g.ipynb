```python
%matplotlib qt
import matplotlib.pyplot as plt
import hyperspy.api as hs
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams["axes.unicode_minus"] = False
```

```python
#load all the datasets

folder = "/home/anderscm/Documents/Data/SPED"


file_1deg ="/Users/anders/Library/CloudStorage/OneDrive-NTNU/Prosjektoppgave/Data/"\
    "SPED/SPED_640x383x12_20x20_9p26x9p26nm_ScanRot20deg_CL12cm_spot0p5nm_alpha5_NBD"\
    "_10ms_1deg_fixed_scan.zspy"

file_0p5deg = "/Users/anders/Library/CloudStorage/OneDrive-NTNU/Prosjektoppgave/Data/"\
    "SPED/SPED_320x192x12_ScanRot20deg_CL12cm_spot0p5nm_alpha5_NBD_10ms_0p5deg.zspy"

file_0deg = "/Users/anders/Library/CloudStorage/OneDrive-NTNU/Prosjektoppgave/Data/"\
    "SED/SED_640x384x12_20x20_9p26x9p26nm_ScanRot20deg_CL12cm_spot0p5nm"\
        "_alpha5_NBD_1ms.zspy"

onedeg = hs.load(file_1deg, lazy=True)
pfivedeg = hs.load(file_0p5deg,lazy = True)
Zerodeg = hs.load(file_0deg, lazy=True)
```

```python
#crop to grain f and g

dp_one = onedeg.inav[290:600,:240]
dp_pfive = pfivedeg.inav[135:295,15:134]
dp_zero = Zerodeg.inav[290:600,:240]
```

```python
#set step size calibration

dp_one.axes_manager[0].scale = 6.92
dp_one.axes_manager[1].scale = 6.92

dp_pfive.axes_manager[0].scale = 13.87
dp_pfive.axes_manager[1].scale = 13.87

dp_zero.axes_manager[0].scale = 6.92
dp_zero.axes_manager[1].scale = 6.92
```

```python
#For the 1deg SPED set
roi_10_4 = hs.roi.CircleROI(cx=-0.00203904, 
                            cy=0.00330131, 
                            r=0.000825327, 
                            r_inner=0)
roi104 = hs.roi.CircleROI(cx=0.00262163,    
                            cy=-0.0035926, 
                            r=0.000922424, 
                            r_inner=0)
bright_field = hs.roi.CircleROI(cx=0.000291292, 
                            cy=-0.000291292, 
                            r=0.000825327, 
                            r_inner=0)


bright_field(dp_one.T).nansum().plot(vmin = 13242, 
                                vmax = 103423, 
                                title = "", 
                                axes_off = True, 
                                colorbar = True, 
                                scalebar = True)
plt.savefig("SPED_1/bf.svg")
plt.close()

roi_10_4(dp_one.T).nansum().plot(vmin = 715, 
                                vmax = 2445, 
                                title = "", 
                                axes_off = True, 
                                colorbar = True, 
                                scalebar = True)
plt.savefig("SPED_1/VDF_neg_104.svg")
plt.close()

roi104(dp_one.T).nansum().plot(vmin = 776, 
                                vmax = 2808, 
                                title = "", 
                                axes_off = True,
                                colorbar = True, 
                                scalebar = True)
plt.savefig("SPED_1/VDF_104.svg")
plt.close()
```
